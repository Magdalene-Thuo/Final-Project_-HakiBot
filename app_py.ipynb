```python
!pip install gradio faiss-cpu sentence-transformers
from google.colab import files
uploaded = files.upload("legal_qa.csv")
# Upload legal_df.pkl and legal_index.index
import pandas as pd
import faiss
from sentence_transformers import SentenceTransformer
import gradio as gr

# Load saved index and data
df = pd.read_pickle("legal_df.pkl")
index = faiss.read_index("legal_index.index")
model = SentenceTransformer('all-MiniLM-L6-v2')

# Define the search function
def get_best_answer(user_input):
    if not user_input.strip():
        return "Please enter a question."
    query_embedding = model.encode([user_input])
    D, I = index.search(query_embedding, k=1)
    return df.iloc[I[0][0]]['answer']

# Create the Gradio interface
iface = gr.Interface(
    fn=get_best_answer,
    inputs=gr.Textbox(lines=2, placeholder="Ask about your rights in Kenya..."),
    outputs="text",
    title="⚖️ HakiBot – Legal Rights Assistant",
    description="Ask about labour laws, police conduct, or arrest rights in Kenya. Powered by AI."
)

# Launch it
iface.launch()
```

```python
!pip install faiss-cpu sentence-transformers pandas

import pandas as pd
import faiss
from sentence_transformers import SentenceTransformer

# Load your legal Q&A dataset
df = pd.read_csv("legal_qa.csv")

# Load model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Encode all questions
question_embeddings = model.encode(df['question'].tolist())

# Build FAISS index
index = faiss.IndexFlatL2(question_embeddings.shape[1])
index.add(question_embeddings)

# Save index and DataFrame
faiss.write_index(index, "legal_index.index")
df.to_pickle("legal_df.pkl")
```
